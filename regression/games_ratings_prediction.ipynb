```python
# loading dataset
import pandas as pd

dataset = pd.read_csv('games.csv')
dataset
```

```python
print(dataset.shape) # number of rows & columns
dataset.describe() 
```

```python
# observing the continious data of future dependent variable y
import matplotlib.pyplot as plt

plt.hist(dataset['average_rating'])
plt.show()
```

```python
# data preprocessing 
dataset = dataset[dataset['average_rating'] > 0]
dataset = dataset.dropna(axis = 0)
plt.hist(dataset['average_rating'])
plt.show()
print(dataset.shape)
```

```python
#correlation_matrix
import seaborn as sns

corrmatrix = dataset.corr()
fig = plt.figure(figsize = (20, 14))
sns.heatmap(corrmatrix, vmax = .8, square = True)
plt.show()
```

```python
# dividing dataset into independent & dependent portions
X = dataset.drop(columns = ['id', 'type', 'name', 'bayes_average_rating', 'average_rating'])
y = dataset['average_rating']
```

```python
# splitting independent & dependent portions into train & test set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state = 0)
```

```python
# applying LinearRegression
from sklearn.linear_model import LinearRegression

linear_regressor = LinearRegression().fit(X_train, y_train)
pd.DataFrame(list(zip(y_test, linear_regressor.predict(X_test))), columns = ['original', 'predicted']).head(5)
```

```python
# evaluating LinearRegression Model
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import numpy as np

r2_score = r2_score(y_test, linear_regressor.predict(X_test))
mse = mean_squared_error(y_test, linear_regressor.predict(X_test))
print('r2_score: ' + str(r2_score))
print('mean_squared_error: ' + str(mse))
print('lr accuracy_score: ' + str(linear_regressor.score(X_test, y_test)))
```

```python
# applying RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor

rf_regressor = RandomForestRegressor(n_estimators=100, random_state = 0).fit(X_train, y_train)
rf_regressor.predict(X_test)
```

```python
# evaluating RandomForestRegressor Model
print('mean squared erroe: ' + str(mean_squared_error(y_test, rf_regressor.predict(X_test))))
print('rfr accuracy score: ' + str(rf_regressor.score(X_test, y_test)))
```

```python
# comparing single prediction value of lr & rfr with original
print('RFR: ' + str(rf_regressor.predict(X_test.iloc[500].values.reshape(1, -1))))
print('LR: ' + str(linear_regressor.predict(X_test.iloc[500].values.reshape(1, -1))))
print('ORG: ' + str(y_test.iloc[500]))
```

```python
# plotting LR predictions , LFR predictions & Original values
plt.scatter([i for i in range(1, 31)], y_test[:30], label = 'original')
plt.scatter([i for i in range(1, 31)], linear_regressor.predict(X_test)[:30], color = 'r', label = 'LR prediction')
plt.scatter([i for i in range(1, 31)], rf_regressor.predict(X_test)[:30], color = 'b', label = 'RFR prediction')
plt.legend()
plt.show()
```
